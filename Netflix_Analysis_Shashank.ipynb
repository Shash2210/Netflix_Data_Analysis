```python
data_movies=data[data['Series or Movie']=='Movie']
```

```python
data_series=data[data['Series or Movie']=='Series']
```

```python
data_movies.head()
```

```python
movie_index=pd.Index(range(1,len(data_movies)+1))
data_movies=data_movies.set_index(movie_index)
```

```python
data_movies.head()
```

```python
movie_genre_list=data_movies.Genre.str.split(',')
```

```python
movie_genre_list
```

```python
movie_genre_dict={}
for genres in movie_genre_list:
    for genre in genres:
        if(genre in movie_genre_dict):
            movie_genre_dict[genre]+=1
        else:
            movie_genre_dict[genre]=1
```

```python
movie_genre_dict
```

```python
movie_genre_data = pd.DataFrame(movie_genre_dict.values(),index = movie_genre_dict.keys(),columns = {'Counts of Genres in Movies'})            
```

```python
movie_genre_data.head(10)
```

```python
movie_genre_data.sort_values(by="Counts of Genres in Movies",ascending=False,inplace=True)
```

```python
movie_genre_data.head(10)
```

```python
movie_genre_data.head(10).plot(kind='bar',y="Counts of Genres in Movies")
plt.ylabel('Count of Genres in Movies',color='red')
plt.xlabel('Genres',color='red')
plt.title('Top 10 Genres in Movies',color='brown')
```

```python
movie_genre_data.head(10).plot(kind='pie',y="Counts of Genres in Movies",figsize=(14,14),autopct='%1.2f%%',counterclock=False,startangle=-180)
```
